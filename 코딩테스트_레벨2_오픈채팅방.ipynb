# 다른 사람 풀이
* heapq https://www.daleseo.com/python-heapq/

```python
import heapq as hq
 
def solution(scoville, K):
 
    hq.heapify(scoville)
    answer = 0
    while True:
        first = hq.heappop(scoville)
        if first >= K:
            break
        if len(scoville) == 0:
            return -1
        second = hq.heappop(scoville)
        hq.heappush(scoville, first + second*2)
        answer += 1  
 
    return answer
```

```python
#product
def solution(numbers, target):
    answer = 0
    from itertools import product
    #합차 조합 만들기 product
    pm=list(product([-1,1],repeat=len(numbers)))
    # print(pm)
    gg=[]
    for i in pm:
        # gg = [ n*s for n,s in zip(numbers,i)]
        # gsum=sum(gg)
        gsum=0
        for j in range(len(i)):
            gsum+=(i[j]*numbers[j])

        # print(i,'=', gsum)
        if gsum==target:
            answer+=1
            # print(i,'=', gsum)
        gg.clear()
    return answer
```

```python
def solution(progresses, speeds):
    answer = []

    #작업완료시간은?
    days=[]
    for p,s in zip(progresses, speeds):
        days.append((100-p)//s + int((100-p)%s!=0))#나머지가 있으면 1일로 계산
    print(days)
    days=days[::-1]#뒤집기
    # print(days)
    while days: 
        max_day = days.pop()
        count=1
        while days:
            next = days.pop()
            if max_day >= next:
                count+=1
            else:
                answer.append(count)
                days.append(next) #다시 집어 넣음
                break
        else:
            answer.append(count)
            
    return answer
```

# 괄호 변환
1. 입력이 빈 문자열인 경우, 빈 문자열을 반환합니다. 
2. 문자열 w를 두 "균형잡힌 괄호 문자열" u, v로 분리합니다.    
단, u는 "균형잡힌 괄호 문자열"로 더 이상 분리할 수 없어야 하며,    
v는 빈 문자열이 될 수 있습니다. 
3. 문자열 u가 "올바른 괄호 문자열" 이라면 문자열 v에 대해    
 1단계부터 다시 수행합니다. 
 - 3-1. 수행한 결과 문자열을 u에 이어 붙인 후 반환합니다. 
4. 문자열 u가 "올바른 괄호 문자열"이 아니라면 아래 과정을 수행합니다. 
 + 4-1 빈 문자열에 첫 번째 문자로 '('를 붙입니다. 
 + 4-2. 문자열 v에 대해 1단계부터 재귀적으로 수행한 결과    
  문자열을 이어 붙입니다. 
 + 4-3. ')'를 다시 붙입니다. 
 + 4-4. u의 첫 번째와 마지막 문자를 제거하고, 나머지 문자열의    
  괄호 방향을 뒤집어서 뒤에 붙입니다. 
 * 4-5. 생성된 문자열을 반환합니다.

```python
pair={'(':')',')':'('}
def isCorrect(p):
    while p.find('()')!=-1:
        p=p.replace('()','')
    return not bool(p) #빈문자열이 아니면.

def solution(p):
    answer = ''
    if p!='': #빈문자열이 아닐때 만
        # print(pair,p)
        pp=p.index(pair[p[0]]) #첫괄호에 대응대는 첫 짝의 위치
        # print('pp',pp)
        u,v = p[:pp+1], p[pp+1:]
        # print(u,' | ',v)
        while u.count('(') != u.count(')'): #균형?
            u+=v[0]
            v=v[1:]
            # print('균형 :', u,' | ',v)
        #u가 올바른가?
        if isCorrect(u): #올바름
            answer = u + solution(v)
        else: #올바르지 않음
            # print('올바르지 않음')
            tmp=[]
            for c in u[1:-1]: #앞뒤 제거 후 괄호방향 뒤집기
                tmp.append(pair[c])            
            # answer = '('+ ''.join(tmp) +')'+ solution(v)
            answer = '('+ solution(v) +')'+ ''.join(tmp)
            # print('a=',answer)
            
    return answer
```

# 행렬 테두리 회전하기

```python
def solution(rows, columns, queries):
    answer = []
    #m행n열 만들기
    mn = [ [ columns*(r-1)+c for c in range(1,columns+1)] for r in range(1,rows+1)]
    # for i in mn: print(list(map(lambda x : f'{x:^3}',i)))
    for q in queries:
        #테두리 가져오기
        border=[]
        x1,y1,x2,y2=q
        w,h = y2-y1, x2-x1
        u,d=mn[x1-1][y1-1:y1+w], mn[x2-1][y1-1:y1+w][::-1]
        #컬럼 접근위해
        zmn = list(zip(*mn))
        # print(zmn[0])
        l,r=zmn[y1-1][x1:x2-1][::-1], zmn[y2-1][x1:x2-1]
        #u,d,l,r 확인
        # for i in [('u = ',u),('d = ',d),('l = ',l),('r = ',r)]: print(*i)
        #border로 합치기
        for i in [u,r,d,l]: border.extend(i)
        # print(border)
        #시계방향으로 회전
        rborder=border[-1:]; rborder.extend(border[:-1])
        # print(rborder)
        #위치 바뀐값 중 최소 찾기
        r_min=10000
        for v1,v2 in zip(border,rborder): r_min=min(r_min,min(v1,v2))
        answer.append(r_min)
        #회전을 행렬에 반영.
        # print('w, h =',w,h)
        for c in range(w+1):
            # print('위아래')
            mn[x1-1][y1-1+c]=rborder[c] #u
            mn[x2-1][y1-1+c]=rborder[(-h-c)] #d
        for r in range(1,h):
            # pritn('좌우')
            mn[x1-1+r][y1-1]=rborder[-r] #l
            mn[x1-1+r][y2-1]=rborder[(w+r)] #l
        #반영 결과 보기
        # for i in mn: print(list(map(lambda x : f'{x:^3}',i)))
        # print('==epoch=='*10)
    return answer

#테스트
import time
s=time.time()
for i in [
          [6,6,[[2,2,5,4],[3,3,6,6],[5,1,6,3]]],
          [3,3,[[1,1,2,2],[1,2,2,3],[2,1,3,2],[2,2,3,3]]],
          [100,97,[[1,1,100,97]]]
          ]:
    print(solution(*i))
print('수행시간 :',time.time()-s, '초')
# 6 	6 	[[2,2,5,4],[3,3,6,6],[5,1,6,3]] 	[8, 10, 25]
# 3 	3 	[[1,1,2,2],[1,2,2,3],[2,1,3,2],[2,2,3,3]] 	[1, 1, 5, 3]
# 100 	97 	[[1,1,100,97]] [1]
```

```python
def solution(numbers):
    answer = 0
    from itertools import permutations
    nl=list(numbers)
    # print(nl)
    #조합 가능한 수 만들기
    candi=[]
    for i in range(len(nl)+1):
        # print(list(map(lambda x : ''.join(x), permutations(nl,i+1))))
        candi.extend(list(map(lambda x : ''.join(x), permutations(nl,i+1))))
    candi=sorted(list(set(map(int,candi))))#중복제거
    # print(candi)
    def isPrime(n):
        if n<2:
            return False
        for i in range(2,int(n**0.5)+1):
            if n%i==0:#나누어지면
                return False
        else:
            return True
    #소수 계수
    for i in candi:
        if isPrime(int(i)):
            # print(i,'는 소수')
            answer+=1
    return answer

#테스트
import time
s=time.time()
for i in [
          '17','011'
          ]:
    print(solution(i))
print('수행시간 :',time.time()-s, '초')
# "17" 	3
# "011" 	2
```

```python
#통과
def solution(name):
    import time
    answer = 0
    abc='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    rbc='A'+abc[::-1]
    nl=list(name)
    tl=list('A'*len(nl))
    cnt=0
    idx=0
    while nl != tl:
        # print(nl,tl,'cnt',cnt)#,abc,rbc)
        if nl[idx]!=tl[idx]:#현재 위치 글자가 다르면
            cnt+=min(abc.find(nl[idx]), rbc.find(nl[idx]))
            # print(abc.find(nl[idx]), rbc.find(nl[idx]), min(abc.find(nl[idx]), rbc.find(nl[idx])))
            tl[idx]=nl[idx] #같은 글자로 바꿈
            continue
        else : #현위치 글자가 같으면 다음 위치로 이동
            # print('여기',len(nl)//2+1)
            for i in range(1,len(nl)//2+1): #문자열을 절반 길이만큼 양측 탐색 
                ri = (idx+i) % len(nl) #오른쪽 끝에서 0으로 이동
                li = (idx-i) if not (idx-i)<0 else len(nl)+(idx-i) #왼쪽끝에서 오른쪽으로
                
                # print((ri,li),(nl[ri],tl[ri]),(nl[li],tl[li]))
                                
                if nl[ri]!=tl[ri]: #동일거리일 때 오른쪽으로 이동
                    idx=ri
                    cnt+=i
                    break
                elif nl[li]!=tl[li]:
                    idx=li
                    cnt+=i
                    break

            # print('새 idx',idx)
            # break  
            # time.sleep(1)     
            
    # print(nl,tl,cnt)
    return cnt

#테스트
import time
s=time.time()
for i in [
        'BABAAAAB',
        'JEROEN',
        'JAZ',
        'JAN',
        'ABABAAAAAAABA'
          ]:
    print(solution(i),'\n')
print('수행시간 :',time.time()-s, '초')
#56
#23
```

# [1차] 뉴스 클러스터링

```python
def solution(str1, str2):
    answer = 0
    #소문자로 바꾸고
    s1=str1.lower()
    s2=str2.lower()
    #2글자씩 짝 지어서 모두 영문자인 짝만 남긴다.
    aset=list(filter(str.isalpha, [''.join(x) for x in zip(s1[:-1],s1[1:])]))
    bset=list(filter(str.isalpha, [''.join(x) for x in zip(s2[:-1],s2[1:])]))
    # print(aset,bset)

    if not aset and not bset:#둘다 공집합이면 자카드 유사도는 1
        answer=1
    else:
        hap=list()
        gyo=list()
        while aset and bset:
            x=aset.pop()
            if x in bset:#bset에도 있으면
                gyo.append(x) #교집합에 더하고
                bset.pop(bset.index(x))#bset에서 제거
            else:
                hap.append(x)
        hap=[ *hap,*gyo,*aset,*bset] #나머지 추가로 더함?
        
        # print(gyo,hap)
        answer= len(gyo)/len(hap)
    return int(answer*65536)

#테스트
import time
s=time.time()
for i in [
        ('FRANCE','french'),
        ('handshake','shake hands'),
        ('aa1+aa2','AAAA12'),
        ('E=M*C^2','e=m*c^2')
          ]:
    print(solution(*i),'\n')
print('수행시간 :',time.time()-s, '초')
# FRANCE 	french 	    16384
# handshake	shake hands 65536
# aa1+aa2 	AAAA12 	    43690
# E=M*C^2 	e=m*c^2 	65536
```

```python
def solution(brown, yellow):
    # answer = []
    a=int(yellow**.5)
    dl=list()
    #노랑의 모든 약수
    for i in range(1,a+1):
        if yellow%i ==0:
            dl.append(i)
    # print(dl)
    #조건에 맞는 가로 세로 찾기
    for d in dl:
        a,b=(d+2),(yellow//d+2)
        if (brown+yellow)==a*b:
            return [max(a,b),min(a,b)]

#테스트
import time
s=time.time()
for i in [
        (10, 	2) 	,
        (8 ,	1) 	,
        (24, 	24) ,
          ]:
    print(solution(*i),'\n')
print('수행시간 :',time.time()-s, '초')

# 10 	2 	[4, 3]
# 8 	1 	[3, 3]
# 24 	24 	[8, 6]
```
