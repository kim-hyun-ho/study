# 좌충우돌 끝에 찾은 공식
## 멀쩡하지 않은 사각형의 좌표를 일일이 기억하지 않고 공식에 의한 풀이로 매우 빠름

```python
#하루 종일 고생해서 스스로 찾은 해답

from math import gcd

#잘린 멀쩡하지 않은 것 구하는 
def mysol(m,n):
    g=gcd(m,n)
    if(g>1):
        return mysol(m//g,n//g) * g ###아래 코드를 보니  mysol은 결국 한번반 재귀 호출한다 왜? gcd는 한번만 구하면 되니까
    return m+n-1 ###한번 재귀 호출되면 식은 (m//g + n//g -1 ) * g 와 같다.

def solution(m,n):
    #전체에서 멀쩡하지 않은것 빼기
    return m*n-mysol(m,n)

'''프로그래머스 다른이의 코드
from math import gcd
def solution(w,h):
    return w * h - (w/gcd(w, h) + h/gcd(w, h) - 1) * gcd(w, h)
'''

import time
start=time.time()
n=100000000
# n=12345678
# n=6
print(solution(n,n-1))
# print(solution(n,3))
print(solution(10000000,10000000-1))
print('--------------')
print(solution(5,7))
print(solution(5,3))
print(solution(5,8))
print(solution(6,3))
print(solution(30,901))
print('time :', time.time() -start)

# n=3#12**2
# for i in range(n-1,0,-1):
#     dots=(mysol(n,i))
#     print(n,i,':',dots,'=>',n*i-(dots),'=',n*i,'-',(dots))
```

# 리스트 버전

```python
#멀쩡한 사각형 좌표 기억에서 단순 계수로 변형해 보자
'''
리스트 버전
'''
def solution(w,h):
    #정사각일때
    if w==h:return w*h-h

    #아래는 실제 이지만 시간이 걸린다
    l,s = max(w,h), min(w,h)  
    dots=[]
    g=s/l
    count=0
    for x in range(1,l):
        yi=s*x//l #정수로 내림 int()보다 빠름
        if g*x > yi: #정수 아닐 때
            if (x-1,yi) not in dots: dots.append((x-1,yi)) #이건 검사필요있음
            dots.append((x,yi))
            if len(dots)>2: #크기가 보다 크면 2개씩 카운트 > 길이는 3~4 일 듯
                count+=2
                dots=dots[2:]
    count+=len(dots)
    # print(l,s,':',count,'=>',l*s-count,'=',l*s,'-',count)
    return w*h-count

import time
start=time.time(); print(solution(8,12))

n=10000000
# print(solution(n,n-1))

# print(solution(n,2))
print(solution(5,6))

# n=4**2;for i in range(n-1,0,-1): (solution(n,i))
# n=7**2;for i in range(n-1,0,-1): (solution(n,i))

print('time :', time.time() -start)
```

# 리스트를 안 쓰는 버전 2개씩 잘렸는지 판단하기의 개선 최종판

```python
#멀쩡한 사각형 좌표 기억에서 단순 계수로 변형해 보자
'''
리스트 안쓰는 버전
마지막만 기억 왜? 두개씩 추가하니까.
'''
def solution(w,h):
    #정사각일때
    if w==h:return w*h-h

    #아래는 실제 이지만 시간이 걸린다
    l,s = max(w,h), min(w,h)  
    ly=-1
    g=s/l
    count=0
    for x in range(1,l): #while보다는 for in 이 빠름.
        yi=s*x//l #정수로 내림 int()보다 빠름
        if g*x > yi: #정수 아닐 때
            # if not (x-1==lx and yi==ly): lx,ly = x-1,yi; count+=1 #왼쪽
            if ly != yi: count+=1 #왼쪽 yi만 비교해도 가능
            ly = yi; count+=1 #오른쪽
    return w*h-count

import time
start=time.time(); print(solution(8,12))

n=100000000
# n=12345678
# n=9
print(solution(n,3))

print('time :', time.time() -start)
#22초까지 줄임 > 최선인가요? 코테는 더 빠른걸 원한다. 그러니까 공식으로 풀기 원함.
```

# 길찾기 버전

```python
#멀쩡한 사각형
'''
길 찾기 버전
'''
def solution(w,h):
    #정사각일때
    if w==h:return w*h-h

    import math
    from fractions import Fraction
    # from decimal import Decimal
    import numpy as np
    l,s = max(w,h), min(w,h)  
    ly=-1
    # g=Fraction(s,l) # 참 느리다 1억,1억-1 : 13분17초
    g=s/l #1억,1억-1 : 23초
    print(f's={s}, l={l}, g={g}')
    count=0
    #기울기를 이용하여 y축으로 1이 되는 x값을 찾아내어 n.xxx 일 때 n+1의 사각형이 잘려나감을 확인 > 서로 소일 때 단축효과 없음
    #서로 소일 때 길찾기 알고리즘은 결국 공식과 일치?
    # x_1=0
    
    #############
    #길찾기 버전 : 우, 상, 대각우상
    #59초 : 느림
    x,y,count=0,0,1
    while 1:
        yn=(x+1)*g
        if  yn < y+1: #우
            x+=1;count+=1
        elif yn == y+1:#대각우상
            x+=1;y+=1;count+=1
        else:#상
            y+=1;count+=1
        if (x==l and y==s):#종점
            count-=1
            break
    print('count:',count)

    return int(w*h-count)

import time
start=time.time()
print(solution(8,12))

n=10000000-1
# n=12345678
n=6
print(solution(n,n-1))
# # print(solution(n,1))
# # print(solution(n,3))
print(solution(5,7))
print(solution(5,3))
print(solution(5,8))
print(solution(6,3))
print(solution(6,4))

print('time :', time.time() -start)
#22초까지 줄임
```

# 참고용 : 프로그래머스 다른이의 답

```python
def gcd(a,b): return b if (a==0) else gcd(b%a,a)    #유클리드 호제법 최대 공약수 구하기
def solution(w,h): return w*h-w-h+gcd(w,h) 
#전체사각형에서 w, h 방향에서 각각 보이는 못 쓰는 사각형을 뺀 후 중복으로 뺀 사각형을 더했네요.
#중복으로 뺀 것이..규칙이 있네요
import time
start=time.time()
n=100000000
print(solution(n,n-1))
print('time :', time.time() -start)
```
