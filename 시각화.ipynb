```python
x= np.arange(-4.5,5,0.5)
y1=2*x**2
y2=5*x+30
y3=4*x**2+10
plt.plot(x,y1,x,y2,x,y3)
plt.show()
```

```python
x=np.arange(-5,5,.1)
y1=x**2-2
y2=20*np.cos(x)**2

plt.figure(1) #그래프 번호 지정
plt.plot(x,y1)

plt.figure(2) #그래프 번호 지정
plt.plot(x,y2)

plt.figure(1) #그래프 번호 지정
plt.plot(x,y2)

plt.figure(2) #그래프 번호 지정
plt.clf() #2번 그래프 지움
plt.plot(x,y1)

plt.show()
```

```python
#하위 그래프 그리기
#plt.subplot(m,n,p) : mxn 행렬에 p번째 p>=1 인덱스 시작은 1부터
x=np.arange(0,10,0.1)
y1=0.3*(x-5)**2+1
y2=-1.5*x+3
y3=np.sin(x)**2
y4=10*np.exp(-x)+1

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
#확대해서 보고싶을 때 출력 범위 조정
x=np.linspace(-4,4,100)
y1=x**3
y2=10*x**2-2

plt.plot(x,y1,x,y2)
plt.show()
```

```python
#잘 안보이니까 확대해보자.
plt.plot(x,y1,x,y2) #확대전에 일단 그려야.
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
```

```python
x=np.arange(0,5,1)
y1=x
y2=x+1
y3=x+2
y4=x+3
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c')
plt.show()
```

```python
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.')
plt.show()
```

```python
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d') #s: square 사각형 , d: diamond 모양
plt.show()
```

```python
plt.plot(x,y1,'>-r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc') #혼합지정
plt.legend(['data1','data2','data3','data4'])#범례
plt.show()
```

```python
#라벨,제목,격자,범례,문자열 표시
x=np.arange(-4.5,5,0.5)
y=2*x**3
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph title')
# plt.grid(True)
plt.grid() #위와 동일
plt.legend(['data1'], loc='lower right') #loc:위치 지정
plt.show()
```
